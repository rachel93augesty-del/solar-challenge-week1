EDA ANalysis 

```python
import matplotlib.pyplot as plt
import pandas as pd
df_benin['Timestamp'] = pd.to_datetime(df_benin['Timestamp'])
```

Plot GHI, DNI, DHI, Tamb vs. Timestamp

```python
plt.figure(figsize=(15,8))
plt.plot(df_benin['Timestamp'], df_benin['GHI'], label='GHI', alpha=0.7)
plt.plot(df_benin['Timestamp'], df_benin['DNI'], label='DNI', alpha=0.7)
plt.plot(df_benin['Timestamp'], df_benin['DHI'], label='DHI', alpha=0.7)
plt.plot(df_benin['Timestamp'], df_benin['Tamb'], label='Tamb', alpha=0.7)
plt.xlabel('Timestamp')
plt.ylabel('Values')
plt.title('Benin: Solar and Temperature Time Series')
plt.legend()
plt.show()
```

Observe monthly trends

```python
df_benin['Month'] = df_benin['Timestamp'].dt.month
monthly_avg = df_benin.groupby('Month')[['GHI', 'DNI', 'DHI', 'Tamb']].mean()

monthly_avg.plot(kind='bar', figsize=(12,6))
plt.title('Benin: Monthly Average Solar & Temperature')
plt.ylabel('Average Value')
plt.show()
```
