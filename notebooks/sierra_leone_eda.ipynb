```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_sierra = pd.read_csv('../src/data/sierraleone-bumbuna.csv')
df_sierra.head()
```

```python
df_sierra['Timestamp'] = pd.to_datetime(df_sierra['Timestamp'])

import matplotlib.pyplot as plt

fig, axs = plt.subplots(4, 1, figsize=(15, 12), sharex=True)

axs[0].plot(df_sierra['Timestamp'], df_sierra['GHI'], color='orange')
axs[0].set_title('Sierra Leone: GHI vs Timestamp')

axs[1].plot(df_sierra['Timestamp'], df_sierra['DNI'], color='red')
axs[1].set_title('Sierra Leone: DNI vs Timestamp')

axs[2].plot(df_sierra['Timestamp'], df_sierra['DHI'], color='green')
axs[2].set_title('Sierra Leone: DHI vs Timestamp')

axs[3].plot(df_sierra['Timestamp'], df_sierra['Tamb'], color='blue')
axs[3].set_title('Sierra Leone: Ambient Temperature vs Timestamp')

plt.xlabel('Timestamp')
plt.tight_layout()
plt.show()
```
