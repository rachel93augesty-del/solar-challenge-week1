```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_togo = pd.read_csv('../src/data/togo-dapaong_qc.csv')

# Display first few rows
df_togo.head()
```

```python
df_togo['Timestamp'] = pd.to_datetime(df_togo['Timestamp'])
```

Plot the key variables vs. Timestamp:

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(4, 1, figsize=(15, 12), sharex=True)

axs[0].plot(df_togo['Timestamp'], df_togo['GHI'], color='orange')
axs[0].set_title('Togo: GHI vs Timestamp')

axs[1].plot(df_togo['Timestamp'], df_togo['DNI'], color='red')
axs[1].set_title('Togo: DNI vs Timestamp')

axs[2].plot(df_togo['Timestamp'], df_togo['DHI'], color='green')
axs[2].set_title('Togo: DHI vs Timestamp')

axs[3].plot(df_togo['Timestamp'], df_togo['Tamb'], color='blue')
axs[3].set_title('Togo: Ambient Temperature vs Timestamp')

plt.xlabel('Timestamp')
plt.tight_layout()
plt.show()
```
